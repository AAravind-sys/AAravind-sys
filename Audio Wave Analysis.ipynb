```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Load an MP3 file
audio_path = "C:\\Users\\Hxtreme\\Documents\\MP3 converter Pro\\P3_Validation Sample-4_Pump sound_20231107203220.mp3"
```

```python

# Sample signal
signal,sr = librosa.load(audio_path)  
signal_length = len(signal)
```

```python
print(signal)
```

```python

# Calculate the FFT
fft_result = np.fft.rfft(signal)

# Generate the frequency axis
frequency_axis = np.fft.rfftfreq(signal_length)

# Plot the magnitude spectrum
plt.plot(frequency_axis, np.abs(fft_result))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```

```python
def spectral_subtraction(signal, noise_estimate):
    clean_signal = np.abs(np.fft.ifft(np.fft.fft(signal) - noise_estimate))
    return clean_signal
```

```python


# Assuming you have a noise estimate (e.g., from a silent portion of the recording)
noise_estimate = min(np.abs(np.fft.fft(signal)))
clean_signal = spectral_subtraction(signal, noise_estimate)
```

```python
clean_signal_length = len(clean_signal)
```

```python
# Calculate the FFT
fft_result = np.fft.rfft(clean_signal)

# Generate the frequency axis
frequency_axis = np.fft.rfftfreq(clean_signal_length)

# Plot the magnitude spectrum
plt.plot(frequency_axis, np.abs(fft_result))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()
```

```python

# Extract audio features
chroma = librosa.feature.chroma_stft(y=y, sr=sr)
mfccs = librosa.feature.mfcc(y=y, sr=sr)

# Visualize the Chroma feature
plt.figure(figsize=(12, 4))
librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.colorbar()
plt.title("Chroma Feature")
plt.show()

# Visualize the MFCCs
plt.figure(figsize=(12, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title("MFCCs")
plt.show()
```
