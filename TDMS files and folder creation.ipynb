```python
import os
import glob
import shutil
import numpy as np
import datetime
```

```python
root = "D:\\New folder"
```

```python
# Get sorted and unique filenames
filenames = sorted(list(set([d[:27] for d in os.listdir(root)])))
```

```python
filenames
```

```python
# Function to convert date format and add serial number prefix
def filetransfer_function(filenames):
    modified_filenames = []
    for idx, filename in enumerate(filenames):
        # Extract date part
        date_part = filename[-8:]
        # Convert date to the desired format
        formatted_date = datetime.datetime.strptime(date_part, '%y-%m-%d').strftime('%d%b%y')
        # Remove old date and append the new formatted date
        new_filename = filename[:-8] + formatted_date
        # Add serial number prefix
        serial_number = f'{idx + 1:02d}_'#02d is f sting format to add zero to prefix of the idx
        modified_filenames.append(os.path.join("D:\\New folder (2)",(serial_number + new_filename)))
    patternlist=[]
    for datepattern in filenames:
        #Add os.path.join regonization pattern('*XX*')
        patternlist.append(f"{'*'}{datepattern[-8:]}{'*'}")
    return patternlist,modified_filenames #this produce two list first index [0] is patternlist and [1] is modified filenames
    
# Apply the function to filenames
folder_name_and_pattern = filetransfer_function(filenames)
```

```python
folder_name_and_pattern
```

```python
#Creater folders in coresponding location
for items in folder_name_and_pattern[1]:
    os.mkdir(items)
```

```python
# gather all files
for pattern,fol_name in zip(folder_name_and_pattern[0],folder_name_and_pattern[1]):
    allfiles = glob.glob(os.path.join(root, pattern), recursive=True)
    for file_path in allfiles:
        dst_path = os.path.join(fol_name, os.path.basename(file_path))
        shutil.move(file_path, dst_path) 
```
