```python
# current directory csv files
csvs = [x for x in os.listdir(directory_name) if x.endswith('.csv')]
# stats.csv -> stats
fns = [os.path.splitext(os.path.basename(x))[0] for x in csvs]

dic_csv = {}
for i in range(len(fns)):
    dic_csv[fns[i]] = pd.read_csv(csvs[i])
```

```python
dic_csv.keys()
```

```python
freq_amp_dfs_Df0 = []
```

```python
for key, df in dic_csv.items():
    # Define the time interval for slicing
    start_time = df.index[0]
    end_time = df.index[-1]
    time_interval = 2000
    
    
    result_df = pd.DataFrame()

    current_time = start_time
    while current_time <= end_time:
        
        next_time = current_time + time_interval

        # Slice the DataFrame based on the specified time range
        sliced_df = df.iloc[(df.index >= current_time) & (df.index <= next_time)]

        df_Filter = sliced_df.loc[(sliced_df['Frequency'] >= 1.700000) & (sliced_df['Frequency'] <= 2.300000)]

        n=len(df_Filter)
        if n >0:
            freq_amp_dfs_Df0.append(df_Filter)
        else:
            # Handle the case where the DataFrame is empty (e.g., print a message)
            print(f"No data in the time range {current_time} - {next_time}")
        current_time = next_time
    #temp_dic_Sf2[key] = result_df     
```

```python
Df0_max_list = []
```

```python
for i in freq_amp_dfs_Df0:
   df_Filter =  i[(i['Frequency'] >= 1.700000) & (i['Frequency'] <= 2.300000)]  
   Df0_max_list.append(df_Filter.loc[df_Filter['Amplitude'].idxmax()])
```

```python
Df0_df = pd.DataFrame(Df0_max_list)
```

```python
Df0_df.reset_index(inplace=True)
```

```python
Df0_df.info()
```

```python
len(Df0_df)
```

```python
time_list_Df0 = [n for n in np.arange(0,2,(2/len(Df0_df)))]
```

```python
len(time_list_Df0)
```

```python
Df0_df['Diaphragm_life'] = time_list_Df0
```

```python
plt.figure(figsize=(20,5))
plt.xticks(np.arange(np.min(Df0_df['Diaphragm_life']),np.max((Df0_df['Diaphragm_life'])),0.4))
plt.plot(Df0_df['Diaphragm_life'],Df0_df['Amplitude'],marker = '.')
plt.xlabel('Hrs')
plt.ylabel('Df0 Freq. Amplitude')
plt.title('Diaphragm lifetime vs Amplitude')
filename = os.path.join(directory_name, f'T17_P1_20_Df0_FFT.png')
plt.savefig(filename)
plt.close() 
```
