```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import fft,signal,stats
```

```python
import os
roots='D:\\Gear Box failure data_Kaggle\\Gear Data'
```

```python
for directory,__, filename in os.walk(roots):
    for name in filename:
        print((name))
```

```python
dfs=[]
for directory,__, filename in os.walk(roots):
    for name in filename:
        # Get gear health status from the file name
        gear_health=name[0] 
        # Get gear load details from the file name
        load=int(name.split('.')[0][5:]) # while using . split we can split filename.csv in to separate str and from there with 0 index we can access the file name alone
        # With os path join merge file name and folder patha and read in to data frame
        df=pd.read_csv(os.path.join(directory,name)) 
        # Insert gear health as separate column
        df['state']=gear_health
        # Insert load as separate column
        df['load']=load
        # Append individual data frame in to list
        dfs.append(df)
```

```python
# concat list of data frames into single data frame
gear_data=pd.concat(dfs).reset_index().rename(columns={'index':'sample_index'})
```

```python
gear_data.head(88321)
```

```python
sensor_reading=gear_data.melt(id_vars=['sample_index','state','load'],
                              value_vars=['a1','a2','a3','a4'],
                              var_name='sensor',value_name='reading'
                              )
```

```python
#filter the data frame (select a1,a2,a3,a4)
def filter_df(df,state=None,load=None,sensor=None):
    df_st=df[df.state==state] if state is not None else df
    df_lo=df_st[df_st.load==load] if load is not None else df_st
    df_se=df_lo[df_lo.sensor==sensor] if sensor is not None else df_lo
    return df_se
```

```python
a1_df=pd.concat([filter_df(sensor_reading,load=0,sensor='a1'),
                 filter_df(sensor_reading,load=90,sensor='a1')
                 ])
```

```python
y_amplitude=np.abs(fft.rfft(sensor_data))
x_freq = fft.rfftfreq(len(sensor_data),0.03) # 0.03 is 30Hz sampling Hz
```

```python
plt.plot(x_freq,y_amplitude)
plt.show()
```

```python
x_peak_spacing = len(y_amplitude)/x_freq.max()
x_peak_prominence = np.quantile(y_amplitude,0.99)
```

```python
peaks,_=signal.find_peaks(y_amplitude,
                          distance=x_peak_spacing,
                          prominence=x_peak_prominence
                          )
```

```python
fig,ax=plt.subplots(figsize=(25,5))
ax.plot(x_freq,y_amplitude)
for peak in peaks:
    ax.scatter(x=x_freq[peak],y=y_amplitude[peak],c='r',marker='o',s=30)
```

```python
x_freq2,y_amplitude2=signal.welch(sensor_data,fs=30)
plt.plot(x_freq2,y_amplitude2)
```
