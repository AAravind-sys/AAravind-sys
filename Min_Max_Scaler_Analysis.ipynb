```python
# Import Basic Packgaes 
import numpy as np
import pandas as pd
import datetime
import statsmodels as sm
import itertools
import glob
import os
from scipy.signal import argrelextrema
from scipy.signal import find_peaks, peak_widths
from sklearn.preprocessing import MinMaxScaler

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns # advanced vizs
from pandas.plotting import lag_plot

# Ignore Warnings
import warnings
warnings.filterwarnings("ignore")

# Set Color Palettes
palette1 = itertools.cycle(sns.color_palette(palette='Set1'))
palette2 = itertools.cycle(sns.color_palette(palette='Set2'))
```

```python
sf2_1c_df = pd.read_csv('D:\\2 inch Pump CSV Files\\2inch Short Test Cases\\1minFFT _Short Test Cases\\Sf2 CSV File\\T17_P1_20_Sf2.csv')
sf2_1c_df.head()
```

```python
sf2_1c_df.drop(columns=['Unnamed: 0','index','Frequency','Cumulative_Amp'], axis=1, inplace=True)
```

```python

# perform a robust scaler transform of the dataset
minmax_1c = MinMaxScaler()
minmax_sf_1c = minmax_1c.fit_transform(sf2_1c_df)
minmax_sf_1c_df = pd.DataFrame(minmax_sf_1c, columns=['minmax_Amplitude','minmax_Diaphragm_Life'])
minmax_sf_1c_df.head()
```

```python
plt.figure(figsize=(20,5))
plt.yticks(np.arange(np.min(minmax_sf_1c_df['minmax_Amplitude']),1.1,0.1))
plt.xticks(np.arange(np.min(sf2_1c_df['Diaphragm_life']),np.max(sf2_1c_df['Diaphragm_life']),50))
plt.plot(sf2_1c_df['Diaphragm_life'],minmax_sf_1c_df['minmax_Amplitude'],marker = '.')
plt.title('Diaphragm Lifetime Vs Amplitude')
plt.ylabel("Norm. Sf2 Amplitude")
plt.xlabel('Hrs.')
plt.show()
```
