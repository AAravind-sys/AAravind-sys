```python
# Define the time interval for slicing
start_time = data_df.index[0]
end_time = data_df.index[-1]
time_interval = 2000

stddev = [[] for _ in range(level)]
mean = [[] for _ in range(level)]
current_time = start_time
while current_time <= end_time:

    next_time = current_time + time_interval

    # Slice the DataFrame based on the specified time range
    sliced_df = data_df.loc[(data_df.index >= current_time) & (data_df.index <= next_time)]
    
    n = len(sliced_df)
    if n > 0:
        coeffs=pywt.wavedec(sliced_df['P2-Air Supply Pressure'],'db2',level=level,mode='symmetric')
        reconstructed_details = []
        for i in range(0, level+1):
            coeffs_copy = [np.zeros_like(c) for c in coeffs]
            coeffs_copy[i] = coeffs[i]
            reconstructed_detail = pywt.waverec(coeffs_copy, 'db2', mode='symmetric')
            reconstructed_details.append(reconstructed_detail)
        for decom_iter in range(1, level+1):
            decomposed_data_set = reconstructed_details[decom_iter]
            mean[decom_iter - 1].append(np.mean(decomposed_data_set))
            stddev[decom_iter-1].append(np.std(decomposed_data_set))
    else:
        pass
    current_time = next_time
```

```python
mean_dic = {}
stadev_dic = {}
for lsindex in range(0,len(mean)):
    mean_dic['mean_{}'.format(lsindex+1)] = pd.DataFrame(mean[lsindex])
for lsindex in range(0,len(stddev)):
    stadev_dic['stddev_{}'.format(lsindex+1)] = pd.DataFrame(stddev[lsindex])
```
